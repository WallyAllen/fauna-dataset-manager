## 3. Demostración de eliminar_por_identificador

La función `eliminar_por_identificador` elimina un registro específico por su ID.

## 4. Demostración de eliminar_por_lista

La función `eliminar_por_lista` elimina registros que coincidan con una lista de IDs.

```python
# Eliminar registros con id en ['1', '3']
eliminaciones.eliminar_por_lista('ejemplo.csv', 'ejemplo_sin_1_3.csv', 'id', ['1', '3'])

# Mostrar el archivo resultante
print("Archivo después de eliminar id='1' y '3':")
with open('ejemplo_sin_1_3.csv', 'r', encoding='utf-8') as file:
    print(file.read())
```

## 5. Demostración de eliminar_por_condicion

La función `eliminar_por_condicion` elimina registros que cumplan una condición matemática en una columna.

```python
# Eliminar registros donde coordinateUncertaintyInMeters > 100
eliminaciones.eliminar_por_condicion('ejemplo.csv', 'ejemplo_sin_alta_incertidumbre.csv', 'coordinateUncertaintyInMeters', '>', '100')

# Mostrar el archivo resultante
print("Archivo después de eliminar registros con incertidumbre > 100:")
with open('ejemplo_sin_alta_incertidumbre.csv', 'r', encoding='utf-8') as file:
    print(file.read())
```

## 6. Demostración de sanitizar_dataset

La función `sanitizar_dataset` elimina registros inválidos basados en validaciones de coordenadas, fechas, etc.

```python
# Crear un CSV con algunos errores para sanitizar
data_con_errores = [
    ['id', 'eventDate', 'decimalLatitude', 'decimalLongitude', 'countryCode', 'coordinateUncertaintyInMeters'],
    ['1', '2023-01-01', '40.7128', '-74.0060', 'US', '100'],
    ['2', '2023-02-01', '100.0', '-118.2437', 'US', '50'],  # Latitud inválida
    ['3', '2023-03-01', '41.8781', '-87.6298', 'US', '75'],
    ['4', '2023-04-01', '37.7749', '-122.4194', 'XX', '200']  # CountryCode inválido
]

with open('ejemplo_con_errores.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerows(data_con_errores)

print("Archivo 'ejemplo_con_errores.csv' creado con datos que incluyen errores.")

# Sanitizar el dataset
eliminaciones.sanitizar_dataset('iadiza', 'ejemplo_con_errores.csv', 'ejemplo_sanitizado.csv')

# Mostrar el archivo sanitizado
print("\nArchivo sanitizado:")
with open('ejemplo_sanitizado.csv', 'r', encoding='utf-8') as file:
    print(file.read())
```
