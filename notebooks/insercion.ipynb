### Bloques de prueba
Estos bloques de código se utilizaron para testear las funciones antes de tener la inserción completa.

```python
from pathlib import Path
import sys
import csv

DIC_BASE = Path.cwd().resolve().parent
sys.path.insert(0, str(DIC_BASE))

from src.insercion import create_record_structure, generate_empty_record, validate_record

file_route = DIC_BASE / 'raw_datasets' / 'inaturalist-filtered' / 'observations.csv'
dataset_name = 'inaturalist'
```

### Ejercicio 4.A y 4.B
Definir la estructura y generar un registro vacío inicializado a partir de las columnas del dataset.

```python
# 4.A: Crear la estructura leyendo el archivo
estructura = create_record_structure(filepath=file_route, id_column='id', delimiter=',')
print("Estructura base del dataset (primeros 5 campos excluyendo ID):")
print(list(estructura.items())[:5])
print("-" * 50)

# 4.B: Generar un registro vacío a partir de una lista de columnas
with open(file_route, encoding='utf-8') as f:
    columnas = csv.DictReader(f, delimiter=',').fieldnames

registro_vacio = generate_empty_record(columns=columnas, id_column='id')
print("Registro vacío generado con generate_empty_record (primeros 5 campos):")
print(list(registro_vacio.items())[:5])
```

### Ejercicio 4.C
Validar un registro antes de insertarlo utilizando `validaciones.py`.


```python
print("Evaluando el registro vacío...")

es_valido = validate_record(registro_vacio, dataset_name=dataset_name)
print(f"¿El registro vacío es válido? {es_valido}")
```
