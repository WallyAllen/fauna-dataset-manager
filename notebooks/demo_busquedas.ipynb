# Demostración del funcionamiento del módulo busquedas

Este notebook muestra cómo usar las funciones `buscar_registros` y `actualizar_registros` del módulo `busquedas.py` para trabajar con archivos CSV.

## 1. Importación de librerías y módulo

Importamos las librerías necesarias y el módulo `busquedas`.

```python
import csv
import os
import sys
sys.path.append('..')
import src.busquedas as busquedas
import src.validaciones as validaciones
```

## 2. Creación de dataset CSV de ejemplo

Creamos un archivo CSV temporal con datos de ejemplo para probar las funciones.

```python
# Datos de ejemplo
data = [
    ['id', 'eventDate', 'decimalLatitude', 'decimalLongitude', 'countryCode', 'coordinateUncertaintyInMeters'],
    ['1', '2023-01-01', '40.7128', '-74.0060', 'US', '100'],
    ['2', '2023-02-01', '34.0522', '-118.2437', 'US', '50'],
    ['3', '2023-03-01', '41.8781', '-87.6298', 'US', '75'],
    ['4', '2023-04-01', '37.7749', '-122.4194', 'US', '200']
]

# Crear archivo CSV
with open('ejemplo.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerows(data)

print("Archivo 'ejemplo.csv' creado con datos de ejemplo.")
```

## 3. Demostración de buscar_registros con filtros

La función `buscar_registros` busca filas en el CSV que cumplan con los filtros especificados.

```python
# Buscar registros con countryCode = 'US'
filtros1 = {'countryCode': 'US'}
resultados1 = busquedas.buscar_registros('ejemplo.csv', filtros1)
print("Resultados para countryCode = 'US':")
for r in resultados1:
    print(r)

print("\n")

# Buscar registros con id = '2'
filtros2 = {'id': '2'}
resultados2 = busquedas.buscar_registros('ejemplo.csv', filtros2)
print("Resultados para id = '2':")
for r in resultados2:
    print(r)

print("\n")

# Buscar con múltiples filtros
filtros3 = {'countryCode': 'US', 'coordinateUncertaintyInMeters': '100'}
resultados3 = busquedas.buscar_registros('ejemplo.csv', filtros3)
print("Resultados para countryCode = 'US' y coordinateUncertaintyInMeters = '100':")
for r in resultados3:
    print(r)
```

## 4. Demostración de actualizar_registros y verificación

La función `actualizar_registros` actualiza un registro específico y guarda los cambios en un nuevo archivo.

```python
# Actualizamos el registro con id = '2', cambiando coordinateUncertaintyInMeters a '50'.
# Usamos dataset='inaturalist' (la estructura del CSV de ejemplo coincide con ese dataset:
# id como identificador, decimalLatitude/decimalLongitude, countryCode, etc., separados por coma).
valores_nuevos = {'coordinateUncertaintyInMeters': '50'}
busquedas.actualizar_registros(
    'ejemplo.csv',
    'ejemplo_actualizado.csv',
    '2',
    'id',
    valores_nuevos,
    dataset='inaturalist',
)

# Verificamos el cambio buscando el registro actualizado
resultados_verif = busquedas.buscar_registros('ejemplo_actualizado.csv', {'id': '2'})
print("Registro actualizado:")
for r in resultados_verif:
    print(r)

# Mostramos todo el archivo actualizado
print("\nArchivo completo actualizado:")
with open('ejemplo_actualizado.csv', 'r', encoding='utf-8') as file:
    print(file.read())
```
